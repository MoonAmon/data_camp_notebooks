### Does average attack and def differ by type of gen I?

- avg of pokemons with only one type

````python
agg_list = ['mean','std','var','count']

# filter the only one type pokemons
poke_df_only_1type = poke_df[:151][poke_df['type_2'].isna()]

poke_df_only_1type.groupby('type_1').agg({'attack':agg_list, 'defense':agg_list}).reset_index().sort_values(ascending=False, by=('attack', 'mean'))
````

- avg of pokemons with two type

````python
poke_df[:151].groupby(['type_1','type_2']).agg({'attack':agg_list,'defense':agg_list}).reset_index().sort_values(ascending=False, by=('attack', 'std'))
````

````python
poke_df[:151][(poke_df['type_1'] == 'Bug') & (poke_df['type_2'] == 'Flying')]
````

### What are the most common types of Pokmon in gen I?

````python
poke_df_genI = poke_df[:151]

type_counts = poke_df_genI['type_1'].value_counts() + poke_df_genI['type_2'].value_counts()

most_common_types_genI = type_counts.sort_values(ascending=False)

most_common_types_genI
````

- Create a diagram bar of the most common types

````python
pyplot.barh(most_common_types_genI.index, most_common_types_genI.values)
````

# Estatística Descritiva Exercícios 

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('6348_tracks_features_2023.csv')
df.info()
```

```python
# Lista de valores nominais
absolute_frequency = df['artists'].value_counts(ascending=False).head(10)
relative_frequency = df['artists'].value_counts(normalize=True).head(10)
cumulative_frequency = absolute_frequency.cumsum()
cumulative_relative_frequency = relative_frequency.cumsum()

frequency_distribution = pd.DataFrame({
    'Absolute Frequency': absolute_frequency,
    'Relative Frequency': relative_frequency,
    'Cumulative Frequency': cumulative_frequency,
    'Cumulative Relative Frequency': cumulative_relative_frequency
})

frequency_distribution
```

```python
frequency_distribution['Absolute Frequency'].plot.barh(color='skyblue',edgecolor='black')
plt.show()
```

```python
r_squared = 0.81
regression_coefficient = 0.33

# Calculate the correlation coefficient
correlation_coefficient = math.sqrt(r_squared)

# Assign the sign of the regression coefficient to the correlation coefficient
if regression_coefficient < 0:
    correlation_coefficient = -correlation_coefficient

print(correlation_coefficient)
```

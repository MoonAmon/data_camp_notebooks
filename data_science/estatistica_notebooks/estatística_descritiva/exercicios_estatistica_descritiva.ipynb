# Exercicios de fixação estatistica descritiva

Utilizando um dataset de dados das musicas que escutei em 2023 do spotify.

```python
import pandas as pd
import json
import re
```

## Limpando dataset

```python
dataset = pd.read_csv('datasets/2000_tracks_features.csv')
dataset
```

```python
album_column = dataset['album'].to_dict()
list_album_names = []
regex_album_name = r"'name': '(.*?)'"
for album in album_column.values():
    list_album_names.append(re.findall(regex_album_name, album))

list_album_names
```

```python
artists_names = []
albums_names = []
for names in list_album_names:
    if len(names) >= 2:
        artists_names.append(names[0])
        albums_names.append(names[-1])
    else:
        artists_names.append(names[0])
        albums_names.append('Unknwon')

print(f"len albums: {len(albums_names)}, len artists: {len(artists_names)}")
```

```python
dataset['album'] = albums_names
dataset['artists'] = artists_names

dataset
```

```python
# Convert 'duration_ms' form milliseconds to minutes
dataset['duration_ms_y'] = dataset['duration_ms_y'] / 60000

# Rename 'duration_ms' to 'duration_min'
dataset.rename(columns={'duration_ms_y':'duration_min'}, inplace=True)

```

```python
dataset.to_csv('6348_tracks_features_2023.csv')
```

# Prbabilidade Anotações

```python
# importar bibliotecas necessarias
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# criar os 2 dados de 6 lados
dice = [1,2,3,4,5,6]
dice2 = dice.copy()

# criar array para os resultados dos dados
results = []

# rodando os dois dados
for _ in range(1000):
    results.append((random.choice(dice), random.choice(dice2)))

results = np.array(results)

sum_results = results.sum(axis=1)

counts = np.bincount(sum_results - 2)

results_dict = {i + 2: count for i, count in enumerate(counts)}


df = pd.DataFrame(list(results_dict.items()), columns=['Resultados','Contagem'])

plt.figure(figsize=(10, 6))
plt.bar(df['Resultados'],df['Contagem'])
plt.grid(True)
plt.plot()
```

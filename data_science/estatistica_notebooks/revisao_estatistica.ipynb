### Normal distribution

```python
data = np.random.normal(0, 1, 10000)
plt.hist(data, bins=30, density=True)
plt.title('Normal Distribution')
plt.show()
```

### Binomial Distribution

```python
# Consider a scenario where we flip a coin 10 times and count the number of heads (successes)

n, p, size = 10, 0.5, 1000
data = np.random.binomial(n, p, size)

plt.hist(data, bins=11, alpha=0.6, color='g')
plt.title('Binomial Distribution')
plt.xlabel('Number of Heads
plt.ylabel('Frquency')
plt.show()
```

### Uniform Distribution

```python
number = 0
data = [number for _ in range(0,10)]

plt.hist(data, bins=1, density=True)
plt.title('Uniform Distribution')
plt.show()
```
